You are a Data Analyst in the Physical Stores Pricing Strategy team, working to ensure competitive pricing for essential household items. Your team is focused on categorizing products by price range and understanding sales trends to maintain affordability for customers. The objective is to use data to identify which price ranges drive the most sales volume, aiding in strategic pricing decisions.

**Question 1 Of 3**
*   What is the total sales volume (i.e. total quantity sold) for essential household items in July 2024? Provide the result with a column named 'Total_Sales_Volume'.

```python
#This question wanna know total sales volumn for essentail
#household items in '2024-07-01' - '2024-07-31'
SELECT SUM(quantity_sold) AS Total_Sales_Volume
FROM fct_sales
WHERE sale_date BETWEEN '2024-07-01' AND '2024-07-31';
```

**Question 2 Of 3**
*   For essential household items sold in July 2024, categorize the items into 'Low', 'Medium', and 'High' price ranges based on their average price. Use the following criteria:

    *   'Low' for prices below $5

    *   'Medium' for prices between \$5 and \$15

    *   'High' for prices above $15

```python
#This question wanna know as each product_name in essentail_household
#was avg item sold(unit_price) in July 2024 and tier that avg into 'Low', 'Medium', 'High'
#base on criteria ,5, 5-15, >15
WITH avg_item_sold AS(
  SELECT d.product_name AS product_name,
    ROUND(AVG(f.unit_price),2) AS avg_item_sold
  FROM fct_sales f
  JOIN dim_products d ON
    f.product_id = d.product_id
  WHERE (f.sale_date BETWEEN '2024-07-01' AND '2024-07-31') AND
    (d.category = 'Essential Household')
  GROUP BY 1
),
tier_item_group AS (
  SELECT product_name,
    CASE
      WHEN avg_item_sold < 5 THEN 'Low'
      WHEN avg_item_sold BETWEEN 5 AND 15 THEN 'Medium'
      WHEN avg_item_sold > 15 THEN 'High'
    END AS categorize_item_follow_criteria
  FROM avg_item_sold
)
SELECT *
FROM tier_item_group
ORDER BY
  CASE categorize_item_follow_criteria
    WHEN 'Low' THEN 1
    WHEN 'Medium' THEN 2
    WHEN 'High' THEN 3
  END DESC;
```

**Question 3 Of 3**
*   Identify the price range with the highest total sales volume for essential household items in July 2024. Use the same criteria as the previous question:

    *  'Low' for prices below $5

    *  'Medium' for prices between \$5 and \$15

    *  'High' for prices above $15
Provide the result with columns named 'Price_Range' and 'Total_Sales_Volume'.

```python
#This question wanna know hightest totalsale from tier segment
#price with own criteria after that sum quantity sold each price range and
#finally show what tier rank with max total sale
#in '2024-07-01' AND '2024-07-31' and category = 'Essential Household'
WITH tier_price_range AS(
  SELECT d.product_name AS product_name,
    CASE
      WHEN f.unit_price < 5 THEN 'Low'
      WHEN f.unit_price BETWEEN 5 AND 15 THEN 'Medium'
      WHEN f.unit_price > 15 THEN 'High'
    END AS Price_Range,
    f.quantity_sold AS quantity_sold
  FROM fct_sales f
  JOIN dim_products d ON
    f.product_id = d.product_id
  WHERE (f.sale_date BETWEEN '2024-07-01' AND '2024-07-31') AND
    (d.category = 'Essential Household')
)
SELECT Price_Range,
  SUM(quantity_sold) AS Total_Sales_Volume
FROM tier_price_range
GROUP BY 1
ORDER BY 2 DESC
LIMIT 1;
```
